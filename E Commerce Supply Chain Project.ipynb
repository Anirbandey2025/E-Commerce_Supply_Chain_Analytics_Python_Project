## **Average Shipping time (days) by Shipping carrier**

```python
avg_ship_time = df.groupby("Shipping carrier")["Shipping time (days)"].mean().sort_values()

print("Average Shipping Time (Days) by Carrier:\n")
print(avg_ship_time.round(3))
```

```python
avg_ship_time.plot(kind="bar", color="skyblue")
plt.title("Average Shipping Time by Carrier")
plt.ylabel("Days")
plt.xticks(rotation=45)
plt.ylim(4.95, 5.02)   # zoom in to highlight differences

# Add value labels
for i, val in enumerate(avg_ship_time):
    plt.text(i, val, f"{val:.3f}", ha='center', va='bottom', fontsize=9)

plt.show()
```

## **Customer Gender Distribution**

```python
# Calculate gender distribution
gender_counts = df["Customer Gender"].value_counts()

print("Customer Gender Distribution:\n")
for gender, count in gender_counts.items():
    pct = (count / gender_counts.sum()) * 100
    print(f"{gender}: {count} customers ({pct:.2f}%)")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

# explicitly set hue to suppress FutureWarning
ax = sns.countplot(x="Customer Gender", hue="Customer Gender",
                   data=df, palette="pastel", legend=False)

plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=9)

# 👇 Zoom y-axis slightly to show tiny differences clearly
ymin = gender_counts.min() * 0.98
ymax = gender_counts.max() * 1.02
plt.ylim(ymin, ymax)

plt.tight_layout()
plt.show()
```

## **Correlation Analysis Between Key Supply Chain Metrics**

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include="number")

# Correlation matrix
corr = numeric_df.corr().round(2)

print("Correlation Summary (Top Relationships):\n")
# Show top 10 absolute correlations (excluding self-correlations)
corr_unstacked = (
    corr.unstack()
    .sort_values(ascending=False)
    .drop_duplicates()[1:11]
)
print(corr_unstacked)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")

plt.title("Correlation Heatmap of Supply Chain Metrics")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## **Shipping Cost by Supplier Location**

```python
# Make sure column is numeric
df["Shipping cost"] = pd.to_numeric(df["Shipping cost"], errors="coerce")

# Calculate average shipping cost by supplier location
cost_by_location = (
    df.groupby("Supplier Location")["Shipping cost"]
      .mean()
      .sort_values(ascending=False)
      .round(2)
)

print("Average Shipping Cost by Supplier Location:\n")
print(cost_by_location)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assign distinct colors for each bar
colors = plt.cm.tab20(np.linspace(0, 1, len(cost_by_location)))

plt.figure(figsize=(9,5))
ax = cost_by_location.plot(kind="bar", color=colors)

plt.title("Average Shipping Cost by Supplier Location")
plt.xlabel("Supplier Location")
plt.ylabel("Average Shipping Cost")
plt.xticks(rotation=45)

# Add value labels on bars
for i, val in enumerate(cost_by_location):
    plt.text(i, val, f"{val:.2f}", ha='center', va='bottom', fontsize=9)

# Slight y-axis zoom for better readability
plt.ylim(cost_by_location.min()*0.98, cost_by_location.max()*1.02)

plt.tight_layout()
plt.show()
```

## **Average Shipping Time by Order Quantity Range**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- Clean numeric data ---
df["Order quantity"] = pd.to_numeric(df["Order quantity"], errors="coerce")
df["Shipping time (days)"] = pd.to_numeric(df["Shipping time (days)"], errors="coerce")
clean_df = df.dropna(subset=["Order quantity", "Shipping time (days)"])

# --- Create bins ---
bins = [0, 20, 40, 60, 80, 100, 120, 140]
clean_df["Qty Bin"] = pd.cut(clean_df["Order quantity"], bins=bins, right=False)

# --- Average per bin ---
avg_by_bin = (
    clean_df.groupby("Qty Bin", observed=False)["Shipping time (days)"]
    .mean()
    .round(2)
)

# --- Drop NaN values (this line kills the posx/posy warning) ---
avg_by_bin = avg_by_bin.dropna()

# --- Plot ---
plt.figure(figsize=(8,4))
ax = avg_by_bin.plot(kind="bar", color="cornflowerblue")

plt.title("Average Shipping Time by Order Quantity Range")
plt.xlabel("Order Quantity Range")
plt.ylabel("Average Shipping Time (days)")
plt.xticks(rotation=0)

# --- Safe value labels ---
for i, v in enumerate(avg_by_bin):
    if np.isfinite(v):   # ✅ only label valid bars
        plt.text(i, v, f"{v:.2f}", ha="center", va="bottom", fontsize=9)

# --- Slight zoom for readability ---
plt.ylim(avg_by_bin.min()*0.98, avg_by_bin.max()*1.02)

plt.tight_layout()
plt.show()
```
